1. What is the difference between a function and a method in Python?

ANS- The key difference is:

Function: A block of code defined with def that works independently and can be called without being tied to an object.
Example-

```python
def greet(name):
    return f"Hello, {name}!"
print(greet("Alice"))  # Output: Hello, Alice!
```

Method: A function defined inside a class, associated with an object, and usually operates on that object.

Example:

```python
class Greeter:
    def greet(self, name):
        return f"Hello, {name}!"

obj = Greeter()
print(obj.greet("Alice"))
```

6. Explain the concept of generators in Python and how they are defined?

Ans- A generator in Python is a special type of iterator that generates values lazily, one at a time, using the yield keyword. This makes them memory-efficient for handling large data.

How to Define:
Use a function with the yield keyword instead of return.
Example:

```python
def my_generator():
    for i in range(3):
        yield i

gen = my_generator()
print(next(gen))  # Output: 0
print(next(gen))  # Output: 1
print(next(gen))  # Output: 2
```

7. What are the advantages of using generators over regular functions?

Ans- Generators offer these advantages over regular functions:

1. **Memory Efficiency**: Generators yield one value at a time, so they don't need to store the entire dataset in memory.
Example: Instead of storing a million numbers, a generator computes them one at a time.

```python
def generate_numbers(n):
    for i in range(n):
        yield i
```

2. **Lazy Evaluation**: Values are produced only when needed, improving performance for large datasets.
Example: Processing lines from a huge file one by one.


```python
def read_file(file_name):
    with open(file_name) as file:
        for line in file:
            yield line
```

3. **Simpler Code**: Generators simplify writing iterators without managing states explicitly.
Example: Instead of using __iter__ and __next__.

```python
def count_up_to(n):
    yield from range(1, n+1)
```

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

Ans- The difference between map(), reduce(), and filter() in Python is:

1. **map**(): Applies a function to each element of an iterable and returns the modified iterable.

Example:

```python
nums = [1, 2, 3]
result = list(map(lambda x: x * 2, nums))  # Multiplies each number by 2
print(result)  # Output will be [2, 4, 6]
```

2.**filter**(): Filters elements of an iterable based on a condition and returns only those that satisfy it.

Example

```python
nums = [1, 2, 3, 4]
result = list(filter(lambda x: x % 2 == 0, nums))  # This filters the even numbers
print(result)  # Output will be [2, 4]
```

3.**reduce**(): Reduces an iterable to a single value by applying a function cumulatively.
(Requires functools module.)

Example

```python
from functools import reduce

nums = [1, 2, 3, 4]
result = reduce(lambda x, y: x + y, nums)  # Adds all numbers in the list
print(result)  # Output will be 10
```

1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

```python
def sum_of_even_numbers(numbers):
    even_sum = 0
    for number in numbers:
        if number % 2 == 0:  # Check if the number is even
            even_sum += number
    return even_sum

# Example usage:
numbers = [1, 2, 3, 4, 5, 6]
result = sum_of_even_numbers(numbers)
print(f"Sum of even numbers: {result}")
```

```python
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, num):
        if num % i == 0:
            return False
    return True

# Test the function for numbers from 1 to 200
for num in range(1, 201):
    if is_prime(num):
        print(num)
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
class FibonacciIterator:
    def __init__(self, terms):
        self.terms = terms
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.terms:
            fib_number = self.a
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return fib_number
        else:
            raise StopIteration

# Example usage:
fib = FibonacciIterator(10)  # Generate Fibonacci sequence for 10 terms
for num in fib:
    print(num)
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_2(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

# Example usage:
exponent = 5
for power in powers_of_2(exponent):
    print(power)
```

7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file_line_by_line(file_path):
    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line.strip()  # strip removes the newline character
    except FileNotFoundError:
        print(f"Error: The file '{file_path}' was not found.")
    except IOError:
        print(f"Error: An issue occurred while reading the file '{file_path}'.")

# Usage example
file_path = 'example.txt'  # Replace with your file path
for line in read_file_line_by_line(file_path):
    print(line)
```

9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

```python
# List of temperatures in Celsius
celsius_temperatures = [0, 10, 20, 30, 40]

# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# Using map() to apply the conversion to each element in the list
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))

# Display the results
print("Celsius temperatures:", celsius_temperatures)
print("Fahrenheit temperatures:", fahrenheit_temperatures)
```

10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
# Function to filter out vowels from a string
def remove_vowels(input_string):
    vowels = 'aeiouAEIOU'
    # Use filter to remove vowels
    result = filter(lambda x: x not in vowels, input_string)
    # Join the filtered result back into a string
    return ''.join(result)

# Example usage
input_string = "Hello World"
output_string = remove_vowels(input_string)
print("Original String:", input_string)
print("String without vowels:", output_string)
```
